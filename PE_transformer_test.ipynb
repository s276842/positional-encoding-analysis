```python
from transformers import pipeline
import torch
```

```python
from transformers import RobertaTokenizer, RobertaForSequenceClassification, AutoTokenizer

name = "cardiffnlp/twitter-roberta-base-sentiment"
model = RobertaForSequenceClassification.from_pretrained(name).cuda()
tokenizer = RobertaTokenizer.from_pretrained(name)
```

```python
sm = torch.nn.Softmax(dim=-1)
```

```python
LABEL_DICT = {0: 'NEG', 1: 'POS'}
import plotly.graph_objects as go
def plot_scores(scores, sentence, label):
  
  scores = scores.cpu().numpy()

  print('='*80)
  print(sentence)  
  # Create traces
  fig = go.Figure()
  fig.add_trace(go.Scatter(x=list(range(len(scores))), y=scores[:, 0],
                      mode='lines+markers',
                      name='Negative'))
  fig.add_trace(go.Scatter(x=list(range(len(scores))), y=scores[:, 2],
                      mode='lines+markers', name='Positive'))
  fig.add_trace(go.Scatter(x=list(range(len(scores))), y=scores[:, 1],
                      mode='lines+markers',
                      name='Neutral'))

  fig.update_layout(title =LABEL_DICT[label],     width=800,
    height=480)
  fig.update_xaxes(range=(0, 514))
  fig.update_yaxes(range=(0, 1))
  
  return fig
```

```python
def get_padded_scores(model, tokenizer, sent):
  scores = []

  for i in range(model.config.max_position_embeddings - sent_len):
    tok = tokenizer(sent)

    input_ids = tok['input_ids']
    input_ids = torch.tensor([[tokenizer.cls_token_id] + [tokenizer.pad_token_id]*i + input_ids[1:]]).cuda()

    attention_mask = torch.ones(1, len(input_ids)).cuda()
    attention_mask[:, 1:1+i] = 0
      
      
    out = model(input_ids = input_ids, attention_mask=attention_mask)
    out = sm(out.logits[0])
    scores.append(out)
  
  scores = torch.stack(scores)
  return scores
```

```python
%matplotlib inline

from datasets import load_dataset
dataset_name = "sst2"

dataset = load_dataset(dataset_name)
dataset['train'][0]


all_scores = []

with torch.no_grad():
  
  for k, sample in enumerate(dataset['train']):
    if k > 10:
      break
    label = sample['label']
    sent = sample['sentence']
    sent_len = len(tokenizer(sent)['input_ids'])
    scores = get_padded_scores(model, tokenizer, sent)

    all_scores.append(scores)
    fig = plot_scores(scores, sent, label)
    fig.show()

```

```python
for i in range(len(all_scores)):
  s = all_scores[i]
  fig = plot_scores(s, dataset['train'][i]['sentence'],  dataset['train'][i]['label'])
  fig.show()
```

```python
for i in range(len(all_scores)):
  s = all_scores[i]
  fig = plot_scores(s, dataset['train'][i]['sentence'],  dataset['train'][i]['label'])
  fig.show()
```

```python
from datasets import load_dataset
dataset_name = "imdb"

dataset = load_dataset(dataset_name)
dataset['train'][0]


all_scores = []

with torch.no_grad():
  
  
  for k, sample in enumerate(dataset['train']):
    
    if k > 10:
      break

    label = sample['label']
    sent = sample['text']
    sent_len = len(tokenizer(sent)['input_ids'])
    
    if len(sent) < 10 or sent_len > 500:
      continue
  
    scores = get_padded_scores(model, tokenizer, sent)
    all_scores.append(scores)
```

```python
for i in range(len(all_scores)):
  s = all_scores[i]
  fig = plot_scores(s, dataset['train'][i]['text'],  dataset['train'][i]['label'])
  fig.show()
```
